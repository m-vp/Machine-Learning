```python
import pandas as pd

# Function to convert 'Final_Marks' to binary target values and drop the column
def process_final_marks(df):
    # Initialize an empty list to hold the target values
    target = []

    # Iterate over 'Final_Marks' column to classify data into binary target values
    for data in df['Final_Marks']:
        if data < 5:
            target.append(0)  # Assign 0 if 'Final_Marks' is less than 5
        else:
            target.append(1)  # Assign 1 if 'Final_Marks' is 5 or greater

    # Add the target list as a new column 'Target'
    df['Target'] = target

    # Drop the 'Final_Marks' column from the DataFrame
    df = df.drop('Final_Marks', axis=1)
    
    return df


# Process the DataFrame to get the target variable and clean up
df = process_final_marks(df)

# Display the updated DataFrame
print(df)
```

```python
#1q)
import numpy as np

# Function to calculate class means
def calculate_class_means(features, target):
    means = {}
    classes = target.unique()  # Get unique class labels

    # Compute the mean of features for each class
    for cls in classes:
        class_data = features[target == cls]  # Filter data for the current class
        means[cls] = class_data.mean(axis=0)  # Calculate the mean for each feature

    return means

# Function to print class means
def print_class_means(means):
    print("Class Means (Centroids):")
    for cls, mean in means.items():
        print(f"Class {cls}: {mean.round(5).tolist()}")

# Main code
features = df.drop(columns=['Target'])
target = df['Target']

# Calculate class means
means = calculate_class_means(features, target)

# Print class means
print_class_means(means)
```

```python
import numpy as np

# Function to calculate class spreads (standard deviations)
def calculate_class_spreads(features, target):
    spreads = {}
    classes = target.unique()  # Get unique class labels

    # Compute the standard deviation of features for each class
    for cls in classes:
        class_data = features[target == cls]  # Filter data for the current class
        spreads[cls] = class_data.std(axis=0)  # Calculate the standard deviation for each feature

    return spreads

# Function to print class spreads
def print_class_spreads(spreads):
    print("Class Spreads (Standard Deviations):")
    for cls, spread in spreads.items():
        print(f"Class {cls}: {spread.round(5).tolist()}")

# Main code
features = df.drop(columns=['Target'])
target = df['Target']

# Calculate class spreads
spreads = calculate_class_spreads(features, target)

# Print class spreads
print_class_spreads(spreads)
```

```python
from numpy.linalg import norm


def calculate_distance(mean1, mean2):
    return norm(mean1 - mean2)


class_0_mean = means[0]
class_1_mean = means[1]

distance = calculate_distance(class_0_mean, class_1_mean)
print(f"Distance between class 0 and class 1: {distance:.5f}")
```

```python
#2q)
import matplotlib.pyplot as plt
import numpy as np

# Function to plot histogram
def plot_histogram(data, bins, xlabel, ylabel, title):
    plt.hist(data, bins=bins, edgecolor='black')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.show()

# Function to calculate mean and variance
def calculate_statistics(data):
    mean = np.mean(data)
    variance = np.var(data)
    return mean, variance

# Main code
# Plot histogram
plot_histogram(df['Correct_code_and_output'], bins=5, xlabel='Correct_code_and_output', ylabel='Frequency', title='Histogram of Correct_code_and_output')

# Calculate and print mean and variance
mean_final_marks, variance_final_marks = calculate_statistics(df['Correct_code_and_output'])
print(f"Mean of Final_Marks: {mean_final_marks:.5f}")
print(f"Variance of Final_Marks: {variance_final_marks:.5f}")
```

```python
#3q)
# Function to extract feature vector from a row
def get_feature_vector(df, index, columns):
    return df.loc[index, columns].values

# Define the columns to extract
columns = ['Header_and_Main_Declaration', 'Incomprehensible_Code', 
           'Comprehensible_Code_with_logical_errors', 
           'Comprehensible_code_with_syntax_errors', 
           'Correct_code_and_output']

# Extract feature vectors
feature_vector_1 = get_feature_vector(df, 0, columns)
feature_vector_2 = get_feature_vector(df, 5, columns)
```

```python
from scipy.spatial.distance import minkowski
import matplotlib.pyplot as plt
import numpy as np

# Function to calculate Minkowski distances for a range of r values
def calculate_minkowski_distances(vec1, vec2, r_values):
    return [minkowski(vec1, vec2, r) for r in r_values]

# Define the range of r values
r_values = np.arange(1, 11)

# Calculate distances
distances = calculate_minkowski_distances(feature_vector_1, feature_vector_2, r_values)

# Plotting
plt.plot(r_values, distances, marker='o')
plt.title('Minkowski Distance vs. Order (r)')
plt.xlabel('Order (r)')
plt.ylabel('Distance')
plt.grid(True)
plt.xticks(r_values)
plt.show()

# Print distances
for r, dist in zip(r_values, distances):
    print(f"Minkowski Distance with r={r}: {dist:.5f}")
```

```python
#4q)
from sklearn.model_selection import train_test_split

# Function to split data into training and testing sets
def split_data(features, labels, test_size=0.3, random_state=42):
    X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=test_size, random_state=random_state)
    return X_train, X_test, y_train, y_test

# Prepare feature vectors and labels
X = df.drop('Target', axis=1)  # Feature vectors
y = df['Target']

# Perform the data split
X_train, X_test, y_train, y_test = split_data(X, y)

# Verify the split
print("Training feature shape:", X_train.shape)
print("Testing feature shape:", X_test.shape)
print("Training labels shape:", y_train.shape)
print("Testing labels shape:", y_test.shape)
print("X_train:")
print(X_train)
print("\nX_test:")
print(X_test)
print("\ny_train:")
print(y_train)
print("\ny_test:")
print(y_test)
```

```python
#5q)
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

# Function to initialize and train a kNN classifier
def train_knn_classifier(X_train, y_train, n_neighbors=3):
    # Initialize the kNN classifier
    model = KNeighborsClassifier(n_neighbors=n_neighbors)
    
    # Train the classifier using the training set
    model.fit(X_train, y_train)
    
    return model

# Initialize and train the kNN model
model = train_knn_classifier(X_train, y_train)

```

```python
#6q)
from sklearn.metrics import accuracy_score

# Function to evaluate the model's accuracy
def evaluate_model_accuracy(model, X_test, y_test):
    # Calculate accuracy
    accuracy = model.score(X_test, y_test)
    return accuracy

# Evaluate and print the accuracy of the kNN classifier
accuracy = evaluate_model_accuracy(model, X_test, y_test)
print("Accuracy of the kNN classifier (k=3):", accuracy)
```

```python
#7q)
# Function to make predictions using the trained model
def make_predictions(model, X_test):
    # Predict class labels for the test data
    predictions = model.predict(X_test)
    return predictions

# Make predictions and print them
predictions = make_predictions(model, X_test)
print("Prediction for the given vector:", predictions)
```

```python
#8q)
from sklearn.metrics import accuracy_score

# Function to evaluate kNN classifier accuracy for different k values
def evaluate_knn_accuracy(X_train, y_train, X_test, y_test, k_range):
    accuracies = []
    
    for k in k_range:
        # Initialize kNN classifier with current k
        neigh = KNeighborsClassifier(n_neighbors=k)
        
        # Train the classifier
        neigh.fit(X_train, y_train)
        
        # Predict on the test set
        y_pred = neigh.predict(X_test)
        
        # Calculate and store accuracy
        accuracy = accuracy_score(y_test, y_pred)
        accuracies.append(accuracy)
    
    return accuracies

# Range of k values to test
k_range = range(1, 12)

# Get accuracies for different k values
accuracies = evaluate_knn_accuracy(X_train, y_train, X_test, y_test, k_range)

# Plotting the accuracy for varying k
plt.figure(figsize=(10, 6))
plt.plot(k_range, accuracies, marker='o')
plt.title('kNN Classifier Accuracy for Varying k')
plt.xlabel('Number of Neighbors (k)')
plt.ylabel('Accuracy')
plt.xticks(k_range)
plt.grid(True)
plt.show()
```

```python
#9q)
model = KNeighborsClassifier(3)
model.fit(X_train, y_train)

from sklearn.metrics import confusion_matrix, classification_report

# Predictions on the training and test set
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Compute confusion matrix
conf_matrix_train = confusion_matrix(y_train, y_train_pred)
conf_matrix_test = confusion_matrix(y_test, y_test_pred)

# Compute classification report
class_report_train = classification_report(y_train, y_train_pred)
class_report_test = classification_report(y_test, y_test_pred)

# Print results
print("Confusion Matrix (Training Set):")
print(conf_matrix_train)

print("\nConfusion Matrix (Test Set):")
print(conf_matrix_test)

print("\nClassification Report (Training Set):")
print(class_report_train)

print("\nClassification Report (Test Set):")
print(class_report_test)
```
